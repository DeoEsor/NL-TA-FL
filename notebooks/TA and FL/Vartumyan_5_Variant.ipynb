* $N = \{ c, d \} $

```python
def check_lang_type(T : list, N : list, P : dict, L : list, S = 'S') -> str:
    """Checks the type of lang and lexic and returns string

    Args:
        T (list): alphabet of lexic terminals
        N (list): alphabet of lexic non-terminals
        P (dict): rules of lang
                  Key: is left part Value: list rights sides
        L (list): alphabet of lang

    Returns:
        str: name of type (this should be enum, but i'm tooooooo lazy for it)
    """

    # P[i] = list
    # P[i][j] = str

    for i in P: #foreach key
      for j in range(len(P[i])):
        if P[i][j] not in T:
            return 'Type 2'
        if P[i][j] not in L:
            return 'Type 3'
        if len(P[i][0])>len(P[i][j]):
            return 'Type 0'

    return 'Type 1'
```
