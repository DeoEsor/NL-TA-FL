```python
import matplotlib.pyplot as plt
```

```python
a_x, a_y  = [0, 2, 8, 13, 15], [0, 0, 0.5, 0, 0]

b_x, b_y = [0, 3, 5, 15], [0, 0, 1, 1]

c_x, c_y = [0, 3, 5, 15], [0, 1, 0, 0]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle('Данные множества')

ax1.plot(a_x, a_y, marker = 'o')
ax2.plot(b_x, b_y, marker = 'o')
ax3.plot(c_x, c_y, marker = 'o')

plt.xlim(0, 15), plt.ylim(0, 1.5)

plt.show()
```

Множество D = !A ∩ B ∪ C , значит, последовательность операций
будет следующей:


1.   !B
2.   !C
3.   A ∩ !B
3.   A ∩ !B ∪ !C

Отдельно множество D

```python
plt.plot(d_x, d_y, marker = 'o')

plt.xlim(0, 15), plt.ylim(0, 1.5)

plt.show()
```

Ядро множества D состоит из интервала (0, 13)



Ядро множества D состоит из элементов из интервала (0,13). Выберем
элемент 5.

 µ A (5) = 0.2

 µ B (5) = 1

 µ C (5) = 0

µ !B (5) = 1 - µ B (5) = 0

µ !C (5) = 1 - µ C (5) = 1

µ A ∩ !B (5) = max{0, µ A (5) + µ !B (5) - 1} = max {0, 0.2 + 0 - 1} = 0

µ A ∩ !B ∪ !C = min {1,  µ A ∩ !B (5) +  µ !C (5)} = min {1,  0 + 1} = 1



```python
plt.plot(d_x, d_y, [5], [1], marker = 'o')

plt.xlim(0, 15), plt.ylim(0, 1.5)

plt.show()
```
